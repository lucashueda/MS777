### 1. Import libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2. Reading datasets

#### Auxiliar array with sheets names

```python
planilha = [str(1997+w) for w in range(0,20)]
planilha
```

### For each excel file (111,222,265 and 333) we create a dataframe

If there is a string named "Unnamed" it is droped


```python
data_111 = []
for p in planilha:
    name =p+'.xlsx'
    data_111.append(pd.read_excel(name, sheetname="111"))
```

```python
for i in range(len(planilha)):
    to_drop = []
    for var in data_111[i].columns:
        if("Unnamed" in var):
            to_drop.append(var)
    data_111[i] = data_111[i].drop(to_drop, axis=1) 
```

```python
data_222 = []
for p in planilha:
    name =p+'.xlsx'
    data_222.append(pd.read_excel(name, sheetname="222"))
```

```python
for i in range(len(planilha)):
    to_drop = []
    for var in data_222[i].columns:
        if("Unnamed" in var):
            to_drop.append(var)
    data_222[i] = data_222[i].drop(to_drop, axis=1) 
```

#### There is no years above 2013 in 333 excel file so it was limited to the year of 2012


```python
data_333 = []
for p in planilha:
    if(int(p)<2013):
        name =p+'.xlsx'
        print(name)
        data_333.append(pd.read_excel(name, sheetname="333"))
```

```python
for i in range(len(planilha)-4):
    to_drop = []
    for var in data_333[i].columns:
        if("Unnamed" in var):
            to_drop.append(var)
    data_333[i] = data_333[i].drop(to_drop, axis=1) 
```

```python
data_265 = []
for p in planilha:
    name =p+'.xlsx'
    data_265.append(pd.read_excel(name, sheetname="265"))
```

```python
for i in range(len(planilha)):
    to_drop = []
    for var in data_265[i].columns:
        if("Unnamed" in var):
            to_drop.append(var)
    data_265[i] = data_265[i].drop(to_drop, axis=1) 
```
