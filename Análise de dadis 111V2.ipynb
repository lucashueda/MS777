### Função que realiza o teste de normalidade em todas as principais variáveis

```python
from scipy import stats
def parametric_test(data,var, method='shapiro', display=True):
    base_aux = pd.DataFrame()
    
    base_aux['Variavel'] = var
    
    statistic = []
    p_value = []
    
    print('O teste escolhido foi ', method)
    
    for v in var:
        if(method == 'shapiro'):
            statistic.append(stats.shapiro(data[v])[0])
            p_value.append(stats.shapiro(data[v])[1])
        elif(method == 'ks'):
            statistic.append(stats.kstest(data[v], cdf='norm')[0])
            p_value.append(stats.kstest(data[v], cdf='norm')[1])
        elif(method == 'anderson'):
            statistic.append(stats.anderson(data[v], 'norm')[0])
            p_value.append(stats.anderson(data[v], 'norm')[1])
        elif(method == 'kurtosis'):
            statistic.append(stats.mstats.normaltest(data[v])[0])
            p_value.append(stats.mstats.normaltest(data[v])[1])
    
    base_aux['Estatistica'] = statistic
    base_aux['P valor'] = p_value
    
    base_aux
    
    return base_aux
```

```python
parametric_test(base_v4,main_vars,'anderson')
```

```python
parametric_test(base_v4,main_vars,'kurtosis')
```

```python
# Criando variáveis de controle "Semana","Mes"

def get_month(day):
    if((day>0) & (day <= 31)):
        return 1
    elif((day>31) & (day <= 59)):
        return 2
    elif((day>59) & (day <= 90)):
        return 3
    elif((day>90) & (day <= 120)):
        return 4
    elif((day>120) & (day <= 151)):
        return 5
    elif((day>151) & (day <= 181)):
        return 6
    elif((day>181) & (day <= 212)):
        return 7
    elif((day>212) & (day <= 243)):
        return 8
    elif((day>243) & (day <= 273)):
        return 9
    elif((day>273) & (day <= 304)):
        return 10
    elif((day>304) & (day <= 334)):
        return 11
    else:
        return 12
    
def get_week(days):
    return np.floor(days/7)
```

```python
base['Mes'] = base['Dia Juliano'].map(get_month)
base['Semana'] = base['Dia Juliano'].map(get_week)
```

```python
from sklearn.cluster import DBSCAN
from collections import Counter
import gc
gc.collect()
```

```python
#setando listas de iteração
anos = base.Ano.unique()

anos = anos.astype(int)
anos.sort()

mes = base.Mes.unique()
mes = mes.astype(int)
mes.sort()


variaveis = [
'Temperatura do Ar Média (°C)',
'Umidade relativa Média (%)',
'Pressão (hPa)',
'Chuva total do período (mm)',
'Veloc. Instantânea do Vento a 2 metros (m/s)',
'Veloc. Instantânea do Vento a 5 metros (m/s)'
]
```

```python
#checando ordenação das variaveis
variaveis
```

```python
# Imports para criar barra de progresso e poder acompanhar o progresso da função
from ipywidgets import FloatProgress
from IPython.display import display
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
#Variáveis de resultado
```

```python
#definindo uma função não genérica para alterar somente sobre qual variável o código irá rodar

def run_dbscan(i):
    
    max_count = len(anos)

    f = FloatProgress(min=0, max=max_count)
    print(f"\tProgresso da função na variável {variaveis[i]}")
    display(f)
    
    outliers_count = 0
    idx_outliers = []

    for a in anos:
        base_aux = base[base.Ano == a]
        for m in mes:
            # pegando base somente no ano em questão
            base_aux2 = base_aux[base_aux.Mes == m]

            if(base_aux2.shape[0] > 0):

                #         Intanciando um modelo DBSCAN
                db = DBSCAN(n_jobs=-1)

                #         Treinando o modelo na variável
                db.fit(base_aux2[variaveis[i]].values.reshape(-1,1))

                outliers_count = outliers_count + (db.labels_ == -1).sum()

                base_aux2['cluster'] = db.labels_
                idx_outliers.extend(base_aux2[base_aux2.cluster == -1].index.values)

        gc.collect()
        f.value += 1

    print("\nConsolidado:")

    print(f"{outliers_count} observações saíriam desse dbscan, e temos {len(idx_outliers)} indices guardados")

    return idx_outliers
```

```python
idx_var1 = run_dbscan(0)
```

```python
idx_var4 = run_dbscan(3)
```

```python
idx_var5 = run_dbscan(4)
```

```python
idx_var6 = run_dbscan(5)
```
