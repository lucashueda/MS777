# Tratamento de dados "errados" e nulos

```python
def fix_year(anos):
    if(anos in [2019, 2039, 2126, 2139]):
        return -999
    else:
        return anos
```

```python
base['Ano'] = base['Ano'].map(fix_year)
```

```python
def fix_hour(hours):
    if(hours == 6640):
        return -999
    else:
        return hours
```

```python
base["Hora - minuto"] = base["Hora - minuto"].map(fix_hour)
```

```python
def fix_day(days):
    if((days<1) | (days > 366)):
        return -999
    else:
        return days
```

```python
base["Dia Juliano"] = base["Dia Juliano"].map(fix_day)
```

```python
base = base.dropna();
```

## 1.2 - Analisando limites de confiança utilizando método Gaussiano e Método de Tukey

```python
vars_analise = base.drop(['Cod', 'Ano', 'Dia Juliano', 'Hora - minuto'], axis = 1).columns
```

```python
base = base[base!=-1000]
```

```python
for data in Anos:
    
    base_data = base[base.Ano == data]
    
    print("O ano é ", data)
    
    for var in vars_analise:

        print(f"A variância da variável {var} é de {base_data[var].var()}")
```

É possível ver que algumas variáveis possuem variâncias muito diferentes por ano de observação, por isso os métodos de outliers serão aplicados ano a ano

```python
from collections import Counter
def Nsigma_method(data,n,features,var_temp,k=3):
    outliers_index = []
    resp_vet = []
    
    dates = data[var_temp].unique()
    dates.sort()
    
    
    for date in dates:
        data_aux = data[data[var_temp] == date]

        for var in features:
            aux_vet = []

            mean_var = data_aux[var].mean()
            std_var = data_aux[var].std()
            
            aux_vet.append(date)
            aux_vet.append(var)
            aux_vet.append(mean_var - n*std_var)
            aux_vet.append(mean_var)
            aux_vet.append(mean_var + n*std_var)
            
            outlier_list_col = data_aux[(data_aux[var] < mean_var - k*std_var) | (data_aux[var] > mean_var + k*std_var )].index
            
            pct_leave = 100*len(outlier_list_col)/data_aux.shape[0]
            pct_leave = round(pct_leave, 2)
            aux_vet.append(pct_leave)
            
            outliers_index.extend(outlier_list_col)
            
            resp_vet.append(aux_vet)
    
    outliers_index = Counter(outliers_index)
    
    result =[]
    
    #Checa um número máximo de outliers em diferentes colunas para se considerar uma coluna a se retirar
    for idx,qt in outliers_index.items():
        if(qt >= n):
            result.append(idx)
    
    df_resp = pd.DataFrame(resp_vet, columns=['Ano','Variável', 'Média - n*sigma', 'Média', 'Média + n*sigma', '% Retirada da variavel-Ano'])
    
    return df_resp, result
```

```python
from collections import Counter
def Tukey_method(data, n, features, var_temp, k_range = 1.5):
    outliers_index = []
    
    resp_vet = []
    
    dates = base[var_temp].unique()
    dates.sort()
    
    for date in dates:
        data_aux = data[data[var_temp] == date]
        for var in features:
            aux_vet = []
            
            P1 = np.percentile(data_aux[var], 25)

            P3 = np.percentile(data_aux[var], 75)

            IQR = P3 - P1

            outlier_range = IQR*k_range

            outlier_list_col = data_aux[(data_aux[var] < P1 - outlier_range) | (data_aux[var] > P3 + outlier_range )].index

            outliers_index.extend(outlier_list_col)
            
            aux_vet.append(date)
            aux_vet.append(var)
            aux_vet.append(P1 - outlier_range)
            aux_vet.append(P3 + outlier_range)
            
            pct_leave = 100*len(outlier_list_col)/data_aux.shape[0]
            pct_leave = round(pct_leave, 2)
            aux_vet.append(pct_leave)
            
            resp_vet.append(aux_vet)
            
    outliers_index = Counter(outliers_index)
    
    result =[]
    
    #Checa um número máximo de outliers em diferentes colunas para se considerar uma coluna a se retirar
    for idx,qt in outliers_index.items():
        if(qt >= n):
            result.append(idx)
        
    df_resp = pd.DataFrame(resp_vet, columns=['Ano','Variável', '1 Quartil - OutRange', '3 Quartil + OutRange', '% Retirada da variavel-Ano'])
    
    return df_resp, result
```

```python
resumo_tukey, to_drop_tukey = Tukey_method(base, 1 , vars_analise, 'Ano', 1.5)
```

# Resumo de variáveis das bases

```python
def resumo_base(data, features):
    results = []
    
    for var in features:
        aux_vet = []
        aux_vet.append(var)
        aux_vet.append(data[var].mean())
        aux_vet.append(data[var].std())
        aux_vet.append(data[var].min())
        aux_vet.append(data[var].max())
        
        results.append(aux_vet)
    
    
    df_results = pd.DataFrame(results, columns=['Variável', 'Média', 'Desvio Padrão', 'Mínimo', 'Máximo'])
    
    return df_results
```
